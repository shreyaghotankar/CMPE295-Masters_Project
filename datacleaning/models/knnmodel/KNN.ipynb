# Matching table for upper and lower clothing items

```python
df_mapping = pd.DataFrame({'top': pd.Series(range(0,100)), 'bottom': pd.Series(range(0,100))})
df_mapping
```

```python
X = df_upper.iloc[:,:][0:100]
Y = df_lower.iloc[:,:][0:100]
Y
```

```python
Y = df_lower.iloc[df_mapping[0:100]["bottom"], :][0:100]
# bottom
Y
```

```python
#lower
X
```

# Splitting the data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
# y_test
# X_test
# y_train
# y_test
```

# Training using KNN

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```
